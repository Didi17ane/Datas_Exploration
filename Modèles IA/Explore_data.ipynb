# **NETTOYAGE DES DONNEES**

## **Traitement des valeurs manquantes** 

```python
val = "Non_connu"

df_ehcvm["branch"] = df_ehcvm["branch"].fillna(val)
df_ehcvm["sectins"] = df_ehcvm["sectins"].fillna(val)
df_ehcvm["csp"] = df_ehcvm["csp"].fillna(val)
```

```python
df_ehcvm.isna().sum()
```

```python
df_ehcvm
```

## **Transforme variable marié** 

```python
val_mar = "Marié(e)"

df_ehcvm["mstat"] = df_ehcvm["mstat"].replace(["Marié(e) polygame", "Marié(e) monogame"], val_mar)
```

```python
df_ehcvm
```

## **Nettoyage variable logement** 

```python
df_ehcvm["logem"].unique()
```

```python
df_ehc = df_ehcvm[df_ehcvm["logem"] == "9"]
df_ehc
```
