## **CRÉATION DE NOUVELLES VARIABLES**

```python
df_ehcvm["sectins"].unique()
```

```python
df_ehcvm["csp"].unique()
```

```python
df_ehcvm["branch"].unique()
```

```python
def definir_emploi_formel(row):
    # Formelle si secteur institutionnel clairement formel
    if row['sectins'] in [
        "Etat/Collectivités locales",
        "Entreprise publique/ parapublique",
        "Entreprise Privée",
        "Organisme international /Ambassade"
    ]:
        return 1

    # Informelle par nature
    if row['sectins'] in [
        "Non_connu",
        "Entreprise associative",
        "Ménage comme employeur de personnel domestique"
    ]:
        return 0

    # Si le secteur institutionnel n’est pas discriminant, utiliser CSP et branche
    if row['csp'] in [
        "Cadre supérieur",
        "Cadre moyen/agent de maîtrise",
        "Ouvrier ou employé qualifié"
    ]:
        return 1
    elif row['csp'] in [
        "Non_connu",
        "Travailleur pour compte propre",
        "Travailleur Familial contribuant pour une entreprise familial",
        "Ouvrier ou employé non qualifié",
        "Manœuvre, aide ménagère",
        "Stagiaire ou Apprenti rénuméré",
        "Stagiaire ou Apprenti non rénuméré",
    ]:
        return 0

    # Ajustement selon la branche si besoin
    if row['branch'] in ["Non_connu", "Agriculture", "Elevage/syl./peche", "Commerce"]:
        return 0
    else:
        return 1

# Application sur ton DataFrame
df_ehcvm['emploi'] = df_ehcvm.apply(definir_emploi_formel, axis=1)

pd.crosstab(df_ehcvm['emploi'], df_ehcvm['empl_formel'], normalize='index') * 100
```
