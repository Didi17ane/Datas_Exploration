```python
new_df = df_tyli.drop(['civility', 'children_number', 'address', 'country', 'city', 'municipality',
       'neighborhood', 'activity_sector_id', 'monthly_charge', 'debt',
       'professional_experience', 'previous_dwellings', 'persons_dependents',
       'created_at', 'updated_at'], axis=1)
new_df
```

```python
new_df['logement']='Locataire'
new_df
```

```python
secteur = new_df[new_df['contract_type'] == "Informel"]
secteur
```

```python
val = "Formel"

new_df["contract_type"] = new_df["contract_type"].replace(['Interim', 'CDI', 'Independant', 'CDD', 'Etudiant', 'Fonctionnaire'], val)
new_df
```

```python
### Calcul de l'âge
new_df["birthday"] = pd.to_datetime(new_df["birthday"])

# Calcul vectorisé de l'âge
today = pd.Timestamp.today()
new_df["age"] = (
    today.year - new_df["birthday"].dt.year
    - ((today.month < new_df["birthday"].dt.month) |
       ((today.month == new_df["birthday"].dt.month) &
        (today.day < new_df["birthday"].dt.day)))
)

# Résultat
print(new_df[["id", "birthday", "age"]])
```

```python
# Définition des tranches d’âge
bins = [0, 17, 24, 34, 44, 54, 64, float("inf")]
labels = ["0-17", "18-24", "25-34", "35-44", "45-54", "55-64", "65+"]

new_df["tranche_age"] = pd.cut(new_df["age"], bins=bins, labels=labels, right=True, include_lowest=True)

# Résultat
print(new_df[["birthday", "age", "tranche_age"]])
```
