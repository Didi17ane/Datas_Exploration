```python
# Choisir le cluster à analyser
cluster_to_analyze = 0
for i in range(97):
    sub = df_100[df_100['cluster'] == cluster_to_analyze]
    #print("_____________________________________________________________________________")
    print(f"Analyse descriptive pour le cluster {cluster_to_analyze} (taille={len(sub)})")

    # Variables numériques
    numeric_vars = ['age_num']
    
    for var in numeric_vars:
        print(f"\nVariable numérique : {var}")
        print(f"  - Médiane : {sub[var].median():.2f}")
        print(f"  - Moyenne : {sub[var].mean():.2f}")
        print(f"  - Écart-type : {sub[var].std():.2f}")
        print(f"  - Minimum : {sub[var].min()}")
        print(f"  - Maximum : {sub[var].max()}")
    
    # Variables catégorielles
    categorical_vars = ['sex', 'marital_status', 'city', 'milieu_resid', 'region_name', 'bancarise']
    
    for var in categorical_vars:
        print(f"\nVariable catégorielle : {var}")
        counts = sub[var].value_counts(dropna=False)
        totals = counts.sum()
        for cat, count in counts.items():
            pct = count / totals * 100
            print(f"  - {cat} : {count} ({pct:.2f}%)")
        mode_val = sub[var].mode(dropna=True)
        if not mode_val.empty:
            print(f"  - Modalité dominante : {mode_val.iloc[0]}")
    print("_____________________________________________________________________________\n")
    cluster_to_analyze = i+1
```

##### **Modalités dominantes**

```python
# Stocker les tableaux pour tous les clusters
results = []

# Listes de variables
numeric_vars = ["age_num"]
categorical_vars = ["sex", "marital_status", "city", "milieu_resid", "region_name", "bancarise"]

# Boucle sur les 97 clusters
for k in range(97):
    sub = df_100[df_100["cluster"] == k]
    
    # ---- Variables numériques ----
    for var in numeric_vars:
        results.append({
            "cluster": k,
            "variable": var,
            "type": "numérique",
            "moyenne": sub[var].mean(),
            "médiane": sub[var].median(),
            "écart_type": sub[var].std(),
            "min": sub[var].min(),
            "max": sub[var].max(),
            "modalité": None,
            "modalité_pct": None
        })
    
    # ---- Variables catégorielles ----
    for var in categorical_vars:
        counts = sub[var].value_counts(normalize=True) * 100
        mode_val = sub[var].mode().iloc[0] if not sub[var].mode().empty else None
        mode_pct = counts.loc[mode_val] if mode_val in counts.index else None
        
        results.append({
            "cluster": k,
            "variable": var,
            "type": "catégorielle",
            "moyenne": None,
            "médiane": None,
            "écart_type": None,
            "min": None,
            "max": None,
            "modalité": mode_val,
            "modalité_pct": mode_pct
        })

# Convertir en tableau final
df_clusters_summary = pd.DataFrame(results)

# Afficher
df_clusters_summary.to_csv('../DATAS/Modalités_Clusters.csv', index=False)
df_clusters_summary
```

##### **Distribution Full Profiles**

```python
def pct_true(s):
    if s.dtype == "bool":
        return float(np.mean(s)) * 100.0
    try:
        return float(np.mean(s.astype(float))) * 100.0
    except Exception:
        return np.nan

# Variables numériques
numeric_vars = ['age_num']

# Variables catégorielles
categorical_vars = ['sex', 'marital_status', 'city', 'milieu_resid', 'region_name', 'bancarise']

profiles = []

for c in sorted(df_100['cluster'].unique()):
    sub = df_100[df_100['cluster'] == c]
    row = {
        'cluster': c,
        'taille': len(sub),
    }
    # Statistiques numériques
    for var in numeric_vars:
        row[f'{var}_median'] = sub[var].median()
        row[f'{var}_mean'] = sub[var].mean()
        row[f'{var}_std'] = sub[var].std()
        row[f'{var}_min'] = sub[var].min()
        row[f'{var}_max'] = sub[var].max()

    # Répartition pour catégorielles
    for var in categorical_vars:
        counts = sub[var].value_counts(dropna=False)
        total = counts.sum()
        repartition = {f'{var}_{cat}': count for cat, count in counts.items()}
        repartition_pct = {f'{var}_{cat}_pct': 100*count/total for cat, count in counts.items()}

        row.update(repartition)
        row.update(repartition_pct)

        # Modalité dominante
        mode_val = sub[var].mode(dropna=True)
        row[f'{var}_mode'] = mode_val.iloc[0] if not mode_val.empty else np.nan

    profiles.append(row)

profiles_df = pd.DataFrame(profiles).sort_values('cluster').reset_index(drop=True)

# Affichage synthétique
with pd.option_context('display.max_columns', None):
    print(profiles_df)

# Option : exporter vers CSV
profiles_df.to_csv("../DATAS/Distribution_full_profiles.csv", index=False)
```
